```python
def prepare_square(df):
    df.loc[df['Square']>(df['Rooms']*40+20),'Square']=df['Rooms']*40+20
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    return df
```

```python
def join_stats1(df,stats1,source_df,mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId','Square_Stat'],how='left')
    #    df=pd.merge(df,stats1,on=['DistrictId','Rooms','HouseYear','Floor'],how='left') - модель переобучилась

    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
data=prepare_square(data)
```

```python
data.loc[:,'Square_Stat'] = pd.Series(np.zeros(len(data['DistrictId'])), index=data.index)
n=4
n_square_stat=10
a=[i*10 for i in range(0,n+int((data['Square'].max())/n_square_stat))]
a
b=[]
j=0
for i in range(1,len(a),n):
    b.append(a[i])
    j+=1
    data.loc[(data['Square']>=a[i-n])&(data['Square']<=a[i]),'Square_Stat']=j-1
data.head(10)
```

```python
stats1=data.groupby(['DistrictId','Square_Stat'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
data=join_stats1(data,stats1,data)
```

```python
features_keys=['Rooms','Square','LifeSquare','KitchenSquare','mean_price','Ecology_2',
               'Ecology_3','Floor','HouseFloor','DistrictId','HouseYear', 'Social_1',
               'Social_2','Social_3','Helthcare_2','Shops_1','Shops_2','Price']
```

```python
train,valid=train_test_split(data.loc[:,features_keys],test_size=0.30, random_state=42)
```

```python
train.shape, valid.shape
```

```python
scaler=StandardScaler()
```

```python
def prepare_kitchen_square(df):
    df.loc[df['KitchenSquare']>(df['LifeSquare']-10),'KitchenSquare']=12
    df.loc[(df['KitchenSquare']<5)&(df['HouseYear']<1990),'KitchenSquare']=6
    df.loc[(df['KitchenSquare']<5)&(df['HouseYear']>=1990),'KitchenSquare']=10
    return df
```

```python
def prepare_life_square(df):
    df.loc[df['LifeSquare']>df['Square'],'LifeSquare']=df['Square']-df['KitchenSquare']-10
    df.loc[df['LifeSquare']<10,'LifeSquare']=df['Square']-df['KitchenSquare']-5
    df.loc[df['LifeSquare']<10,'LifeSquare']=df['Square']-df['KitchenSquare']
    return df
```

```python
def fillna_lifesquare(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].mean())
    #df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].median())
    return df
```

```python
def dummies(df,features='features_keys'):
    df[features]=df[features].astype(object)
    df=pd.get_dummies(df)
    return df
```

```python
features_keys=['Ecology_2','DistrictId'] 
```

```python
train=prepare_kitchen_square(train)
train=prepare_life_square(train)
train=fillna_lifesquare(train,train)
train=dummies(train,features=features_keys)
```

```python
#train['Rooms']=scaler.fit_transform(train.loc[:,['Rooms']])
train['Square']=scaler.fit_transform(train.loc[:,['Square']])
train['LifeSquare']=scaler.fit_transform(train.loc[:,['LifeSquare']])
train['KitchenSquare']=scaler.fit_transform(train.loc[:,['KitchenSquare']])
train['HouseFloor']=scaler.fit_transform(train.loc[:,['HouseFloor']])
train['HouseYear']=scaler.fit_transform(train.loc[:,['HouseYear']])
```
