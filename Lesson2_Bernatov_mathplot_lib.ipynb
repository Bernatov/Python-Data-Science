```python
t=np.linspace(0,10,51)
```

```python
import math
```

```python

f=[math.cos(i) for i in t]
```

```python
plt.plot(t,f, color='green')
plt.title(
    "График функции f(x)",
    fontsize=16,
    fontweight="bold",
    color="black",
    family="serif",
)
plt.xlabel(
    "Значения t",
    fontsize=16,
    fontweight="bold",
    color="black",
    family="serif",
)
plt.ylabel(
    "Значения f",
    fontsize=16,
    fontweight="bold",
    color="black",
    family="serif",
)
plt.axis([0.5,9.5,-2.5,2.5])

plt.show()
```

```python
x=np.linspace(-3,3,51)
```

```python
y1=[i**2 for i in x]
```

```python
y2=[(i*2+0.5) for i in x]
```

```python
y3=[(i*(-3)-1.5) for i in x]
```

```python
y4=[math.sin(i) for i in x]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График x^2")
ax2.plot(x, y2)
ax2.set_title("График x*2+0.5")
ax3.plot(x, y3)
ax3.set_title("График -3x*2-1.5")
ax4.plot(x, y4)
ax4.set_title("График sin(x)")
```

```python
class1=data.loc[data['Class']==1]
class1=class1['V1']
```

```python
class0=data.loc[data['Class']==0]
class0=class0['V1']
```

```python
fig= plt.figure()
ax=fig.add_subplot(111)
ax.hist(class0,density=True, bins=20, alpha=1, ec='black',color='grey')
ax.hist(class1,density=True, bins=20, alpha=0.5, ec='black',color='red')
plt.title("Гистограмма Транзакций")
plt.legend(labels=['Class 0', 'Class 1'])
plt.xlabel("Class")
plt.show()
```
